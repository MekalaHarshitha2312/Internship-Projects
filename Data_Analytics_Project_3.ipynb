```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = pd.read_csv('Iris.csv')
```

```python
#Display the first few rows of the dataset
print(iris_df.head())
```

```python
#Visualize the distribution of features using histograms

iris_df.hist(bins=20, figsize=(10, 8), color='skyblue')
plt.suptitle('Distribution of Features in Iris Dataset', fontsize=16)
plt.show()
```

```python
#You can also visualize the distribution with Seaborn for a more detailed view

plt.figure(figsize=(12, 6))
for i, column in enumerate(iris_df.columns[1:-1], 1):  # Skip 'Id' and 'Species'
    plt.subplot(2, 2, i)
    sns.histplot(iris_df[column], kde=True, color='purple')
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```
